# Descripción y preparación del dataset

Este item contempla el entrenamiento de clasificadores para la predicción de competencias de usuarios de LinkedIn. LinkedIn es una red social orientada a la formación de contactos profesionales y de negocios. Los usuarios en esta red se registran y declaran las competencias, habilidades o capacidades que poseen de forma de captar contactos para iniciar relaciones laborales o de negocio. 

> a) Defina una función que cargue lea el primer archivo y genere una matriz dispersa $Z$ (preferentemente una `scipy.sparse.csr_matrix` si trabaja en python) de forma $n × d$ donde $n$ es el número de perfiles ́LinkedIn (número de ejemplos en nuestro caso) y d es el número de competencias totales. La entrada $Z_{ij}$ será $1$ si el usuario $i$ posee/declara la competencia $j$ y $0$ en otro caso.

Primero que todo se debe realizar la carga del dataset. El dataset se encuentra inicialmente en dos archivos: 
* El archivo de competencias `skill_id` que contiene los id's y los nombres de las competencias
* El archivo de competencias de usuario `user_skills` que contiene los id's de los usuarios y los id's de las competencias que ellos declaran poseer

Los datos son almacenados en una matriz dispersa de tipo `csr_matrix`. Se definen las funciones `load_skills(filename)` y `load_userskills(filename,skillcount)` para cargar los archivos de competencias y competencias de usuario respectivamente. Estas funciones también informan cuantas competencias y usuarios hay respectivamente. 

El dataset contiene 14544 competencias de 7890 usuarios.

```python
import numpy as np
from scipy.sparse import csr_matrix
import random
random.seed(1513) 

def load_skills(filename):
	sifile = open(filename)
	skills = []
	for line in sifile:
		skills.append(line.split(':')[0].decode('utf-8'))

	sifile.close()
	return skills, len(skills)

def load_userskills(filename,skillcount):
	usfile = open(filename)
	data = []
	row_ind = []
	col_ind = []

	usercount = 0
	for line in usfile:
		user = int(line.split(':')[0])
		uskills = [int(i) for i in line.split(':')[1].split(',')]
		for s in uskills:
			data.append(1)
			row_ind.append(usercount)
			col_ind.append(s)
		usercount+=1
	usfile.close()
	return csr_matrix((data,(row_ind,col_ind)),shape=(usercount,skillcount)), usercount
	

skills, skillcount = load_skills('skill_id')
X, usercount = load_userskills('user_skill',skillcount)


print 'Competencias: %d'%skillcount
print 'Usuarios: %d'%usercount
```

# Entrenamiento de los clasificadores

> b) Construya dos sub-matrices, $Z_{tr}$ , $Z_{ts}$, a partir de $Z$. La primera, de dimensiones $n_{tr}×d$ corresponderá a los casos de entrenamiento. La segunda, de dimensiones $n_{ts}×d$ corresponderá a los casos de pruebas. Los conjuntos de entrenamiento y pruebas deben ser disjuntos y además $n_{tr} + n_{ts} = n$. Elija aleatoriamente
los casos de entrenamiento y pruebas, permutando por ejemplo las filas de $X$ antes de la selección.

Se procede a la creación de los conjuntos de entrenamiento y prueba para los clasificadores. El conjunto de entrenamiento considerará un 70% de los ejemplos (usuarios) seleccionados al azar.

```python
import random
random.seed(1513) #para asegurar siempre el mismo resultado y se pueda repetir el experimento

traincount = int(round(usercount*0.7))
testcount = usercount-traincount

train_ind = random.sample(range(0,usercount),traincount)
train_ind.sort()
test_ind = [x for x in range(0,usercount) if x not in train_ind]

Xtrain = X[train_ind,:]
Xtest = X[test_ind,:]

print "Dim train set: %s"%str(Xtrain.shape)
print "Dim test set: %s"%str(Xtest.shape)
```

# Resultados

Al observar la *accuracy* en el conjunto de entrenamiento de los clasificadores se pueden ver resultados bastante buenos en la mayoría de las instancias.

En las instancias que se corresponden con competencias declaradas por muchos usuarios (PHP, Liderazgo de equipos, Microsoft SQL Server) todos los clasificadores obtuvieron una buena *accuracy* con SVM Lineal y Regresión Logística como los mejores clasificadores seguidas por Naive Bayes, k-Nearest Neighbors y SVM con kernel Gaussiano. En estas instancias se nota una clara diferencia entre los clasificadores y los clasificadores Dummy.

Sin embargo, en las instancias con competencias declaradas por pocos usuarios el clasificador Dummy lográ una muy buena *accuracy*. Esto es porque la mayoría de los ejemplos son de clase negativa (igual a cero, el usuario no tiene la competencia). El clasificador dummy sigue siendo inferior en *accuracy* a todos los clasificadores a excepción del SVM con kernel Gaussiano.

```python
from sklearn.naive_bayes import BernoulliNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.dummy import DummyClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import recall_score
from sklearn.metrics import accuracy_score

def benchmark(model, model_name, Xtrain, ytrain, Xtest, ytest):
    model.fit(Xtrain,ytrain)
    print "%s training\t%f\t%f"%(model_name, model.score(Xtrain,ytrain), recall_score(ytrain,model.predict(Xtrain)))
    print "%s testing\t%f\t%f"%(model_name, model.score(Xtest,ytest), recall_score(ytest,model.predict(Xtest)))
    return (model.score(Xtrain,ytrain), recall_score(ytrain,model.predict(Xtrain)),model.score(Xtest,ytest), recall_score(ytest,model.predict(Xtest)))

dums_s = []
dumu_s = []
bnb_s = []
lgr_s = []
svml_s = []
svmg_s = []
knn_s = []

#Notar que algunos clasificadores (SVMs y KNN) demoran en ejecutarse
for itu in idxs:
    print (itu,skills[itu],skill_freq[itu], round(ad[itu],3))
    
    Xtrainc = Xtrain[:,range(0,itu) + range(itu+1,skillcount)]
    ytrainc = Xtrain[:,itu].toarray().transpose()[0]
    Xtestc = Xtest[:,range(0,itu) + range(itu+1,skillcount)]
    ytestc = Xtest[:,itu].toarray().transpose()[0]
    
    print 'Model name\tAccuracy\tRecall'
    
    s = benchmark(DummyClassifier(strategy='stratified'), 'DUMS', Xtrainc, ytrainc, Xtestc, ytestc)
    dums_s.append(s)
    s = benchmark(DummyClassifier(strategy='uniform'), 'DUMU', Xtrainc, ytrainc, Xtestc, ytestc)
    dumu_s.append(s)
    s = benchmark(BernoulliNB(), 'bNB', Xtrainc, ytrainc, Xtestc, ytestc)
    bnb_s.append(s)
    s = benchmark(LogisticRegression(), 'LGR', Xtrainc, ytrainc, Xtestc, ytestc)
    lgr_s.append(s)
    s = benchmark(SVC(kernel='linear',class_weight='balanced'), 'SVMl', Xtrainc, ytrainc, Xtestc, ytestc)
    svml_s.append(s)
    #benchmark(SVC(kernel='poly',class_weight='balanced'), 'SVMp', Xtrainc, ytrainc, Xtestc, ytestc)
    #svm con kernel polinomial daba malos resultados
    s = benchmark(SVC(kernel='rbf',class_weight='balanced'), 'SVMg', Xtrainc, ytrainc, Xtestc, ytestc)
    svmg_s.append(s)
    s = benchmark(KNeighborsClassifier(n_neighbors = 10), 'KNN', Xtrainc, ytrainc, Xtestc, ytestc)
    knn_s.append(s)

    
```

```python
exes = [skills[i] for i in idxs]

n = 0
yplt = [x[n] for x in dums_s]
plt.plot(range(len(exes)), yplt, 'k--', label='Dummy Stratified')
yplt = [x[n] for x in dumu_s]
plt.plot(range(len(exes)), yplt, 'k-.', label='Dummy Uniform')
yplt = [x[n] for x in bnb_s]
plt.plot(range(len(exes)), yplt, 'r-', label='Naive Bayes')
yplt = [x[n] for x in lgr_s]
plt.plot(range(len(exes)), yplt, 'g-', label='Logistic Regression')
yplt = [x[n] for x in svml_s]
plt.plot(range(len(exes)), yplt, 'b-', label='SVM Linear')
yplt = [x[n] for x in svmg_s]
plt.plot(range(len(exes)), yplt, 'c-', label='SVM Gaussian')
yplt = [x[n] for x in knn_s]
plt.plot(range(len(exes)), yplt, 'm-', label='K-Nearest Neighbors')

plt.title('Accuracy Score Training Set')

plt.legend(bbox_to_anchor=(1.05, 1.),loc=2,
           ncol=1, borderaxespad=0.)

plt.xticks(range(len(exes)), exes,rotation=10)
plt.show()
```

Al observar la *accuracy* en el conjunto de pruebas se obtiene un resultado similar. En las instancias con competencias declaradas por muchos usuarios se obtiene un *accuracy* superior al del dummy, pero no hay mucha diferencia en las instancias con pocos usuarios.

# Conclusiones

Se llevó a cabo predicción de competencias para un dataset basado en datos extraídos desde la red social LinkedIn. Se observó que la gran mayoría de las competencias eran declaradas por pocos usuarios. Se entrenaron 5 clasificadores (Bayesiano Ingenuo, Regresión Logística, SVM Lineal, SVM con kernel Gaussiano y k-Nearest Neighbors) para 5 competencias con distinta rareza.

Se observó que los mejores clasificadores, en los casos de competencias comunes, venían a ser SVM lineal y Regresión Logística, pero la mayoría (a excepción de k-Nearest Neighbors) lograron un desempeño decente sobre el conjunto de pruebas.

Los otros clasificadores también vieron una baja considerable en su desempeño. En estos casos el clasificador SVM Gaussiano resultó el mejor para la medida *recall*.

Pareciera ser que estos clasificadores tienen problemas al tener un dataset con un desbalance tan grande entre clases y podría deberse a una dificultad de encontrar relaciones de estas competencias raras con otras (sean raras o comunes).
